```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
import numpy as np
import matplotlib.pyplot as plt
start_time = time.time()
def sort_by_selection(line):
    for i in range(len(line)):
        min_index = i
        temp = line[min_index]
        for j in range(i, len(line)):
            if line[j] < temp:
                temp = line[j]
                min_index = j
        line[min_index] = line[i]
        line[i] = temp


N = []
t = []
n = 100

for m in range(1, 100):
    matrix = matrix_generator(n, m, 0, 9)
    start_time = time.time()
    
    for i in range(0, len(matrix)-1):
        matrix[i] = sort_by_selection(matrix[i])
    
    Time = round((time.time() - start_time)*1000)
    amount = n * m

    N.append(amount)
    t.append(Time)

fig, ax = plt.subplots()

ax.plot(N, t, linewidth = 2.5)
ax.set_xlabel("n")
ax.set_ylabel("t, мс")
ax.set_title('Сортировка выбором')
ax.grid(which='major', color = 'k')

ax.minorticks_on()
ax.grid(which='minor', color = 'gray', linestyle = '-')

plt.show()

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
#сортировка вставкой
start_time = time.time()


def sort_by_insertion(line):
    for i in range(1, len(line)):
        for j in range(i, 0, -1):
            if line[j] < line[j - 1]:
                line[j], line[j - 1] = line[j - 1], line[j]
            else:
                break


N = []
t = []
n = 100

for m in range(1, 100):
    matrix = matrix_generator(n, m, 0, 9)
    start_time = time.time()
    
    for i in range(0, len(matrix)-1):
        matrix[i] = sort_by_insertion(matrix[i])
    
    Time = round((time.time() - start_time)*1000)
    amount = n * m

    N.append(amount)
    t.append(Time)

fig, ax = plt.subplots()

ax.plot(N, t, linewidth = 2.5)
ax.set_xlabel("n")
ax.set_ylabel("t, мс")
ax.set_title('Сортировка вставкой')
ax.grid(which='major', color = 'k')

ax.minorticks_on()
ax.grid(which='minor', color = 'gray', linestyle = '-')

plt.show()

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка обменом. || Сортировка пузырьком.

def bubble_sort(line):
    num_of_swap = 1
    for i in range(len(line) - 1):
        for j in range(len(line) - num_of_swap):
            if line[j] > line[j + 1]:
                line[j], line[j + 1] = line[j + 1], line[j]
        num_of_swap += 1


        
N = []
t = []
n = 100

for m in range(1, 100):
    matrix = matrix_generator(n, m, 0, 9)
    start_time = time.time()
    
    for i in range(0, len(matrix)-1):
        matrix[i] = bubble_sort(matrix[i])
    
    Time = round((time.time() - start_time)*1000)
    amount = n * m

    N.append(amount)
    t.append(Time)

fig, ax = plt.subplots()

ax.plot(N, t, linewidth = 2.5)
ax.set_xlabel("n")
ax.set_ylabel("t, мс")
ax.set_title('Сортировка пузырьком')
ax.grid(which='major', color = 'k')

ax.minorticks_on()
ax.grid(which='minor', color = 'gray', linestyle = '-')

plt.show()


print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))

```

```python
#Сортировка Шелла
start_time = time.time()


def shell_sort(line):
    n = len(line)
    step = n // 2
    while step > 0:
        for i in range(step, n):
            position = i
            while position >= step and line[position] < line[position - step]:
                line[position], line[position - step] = line[position - step], line[position]
                position -= step
        step = step // 2


N = []
t = []
n = 100

for m in range(1, 100):
    matrix = matrix_generator(n, m, 0, 9)
    start_time = time.time()
    
    for i in range(0, len(matrix)-1):
        matrix[i] = shell_sort(matrix[i])
    Time = round((time.time() - start_time)*1000)
    amount = n * m

    N.append(amount)
    t.append(Time)

fig, ax = plt.subplots()

ax.plot(N, t, linewidth = 2.5)
ax.set_xlabel("n")
ax.set_ylabel("t, мс")
ax.set_title('Сортировка Шелла')
ax.grid(which='major', color = 'k')

ax.minorticks_on()
ax.grid(which='minor', color = 'gray', linestyle = '-')

plt.show()
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Быстрая сортировка.(Хоара)
start_time = time.time()


def quick_sort(line):
    if len(line) > 1:
        x = line[random.randint(0, len(line) - 1)] #случайное пороговое значение для разделения на малые и большие
        low = [el for el in line if el < x]
        equal = [el for el in line if el == x]
        high = [el for el in line if el > x]
        line = quick_sort(low) + equal + quick_sort(high)

    return line


N = []
t = []
n = 100

for m in range(1, 100):
    matrix = matrix_generator(n, m, 0, 9)
    start_time = time.time()
    
    for i in range(0, len(matrix)-1):
        matrix[i] = quick_sort(matrix[i])
    Time = round((time.time() - start_time)*1000)
    amount = n * m

    N.append(amount)
    t.append(Time)

fig, ax = plt.subplots()

ax.plot(N, t, linewidth = 2.5)
ax.set_xlabel("n")
ax.set_ylabel("t, мс")
ax.set_title('Быстрая сортировка')
ax.grid(which='major', color = 'k')

ax.minorticks_on()
ax.grid(which='minor', color = 'gray', linestyle = '-')

plt.show()

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))

    
```

```python
#турнирная сортировка
def heapify(line, n, i):
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2

    if l < n and line[i] < line[l]:
        largest = l
    if r < n and line[largest] < line[r]:
        largest = r

    if largest != i:
        line[i], line[largest] = line[largest], line[i]
        heapify(line, n, largest)


def tournament_sort(line):
    N = len(line)
    for i in range(N // 2, -1, -1):
        heapify(line, N, i)
    for i in range(N - 1, 0, -1):
        line[i], line[0] = line[0], line[i]
        heapify(line, i, 0)
    return line


N = []
t = []
n = 100

for m in range(1, 100):
    matrix = matrix_generator(n, m, 0, 9)
    start_time = time.time()
    
    for i in range(0, len(matrix)-1):
        matrix[i] = tournament_sort(matrix[i])
    Time = round((time.time() - start_time)*1000)
    amount = n * m

    N.append(amount)
    t.append(Time)

fig, ax = plt.subplots()

ax.plot(N, t, linewidth = 2.5)
ax.set_xlabel("n")
ax.set_ylabel("t, мс")
ax.set_title('Турнирная сортировка')
ax.grid(which='major', color = 'k')

ax.minorticks_on()
ax.grid(which='minor', color = 'gray', linestyle = '-')

plt.show()

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))                            
```
